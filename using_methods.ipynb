## Applying the transformations

#### This jupyter contains the applications of the methods created in the classes of the proj_oo_etl

```python
from Connection import Connection
from Extract import Extract
from Transform import Transform
```

```python
# Creating an instance of the connection object
instance_con = Connection()
datalake = instance_con.mysql_con('datalake')
dw = instance_con.mysql_con('data_warehouse')
```

```python
# Creating an instance of the extract object
instance_extract = Extract()
```

```python
# Creating a dataframe using a local csv
economic_freedom = instance_extract.create_df_using_csv(
    csv_path_and_file_name\
         = '/home/welbert/Documents/proj_oo/exported_datasets/economic_freedom_test_file.csv',
         separator = ';'
    )
```

```python
# creating df before transformation
teste = economic_freedom[
        ["Population (Millions)", 
        "GDP (Billions, PPP)", 
        "GDP per Capita (PPP)", 
        "Unemployment (%)", 
        "FDI Inflow (Millions)"]]
```

```python
# Printing Information before apply the transformation in the teste df
print(teste.info())
teste.head()
```

```python
# Creating an instance of transform object and transforming the float columns
t = Transform()

economic_freedom_transformed\
    = t.transform_columns_type(
        economic_freedom,
        [ 
            "GDP (Billions, PPP)", 
            "GDP per Capita (PPP)", 
            "Unemployment (%)", 
            "FDI Inflow (Millions)"
        ], 
        'float')

economic_freedom_transformed.head()
```

```python
# creating and printing df after transformation
teste\
     = economic_freedom_transformed[
        ["Population (Millions)", 
        "GDP (Billions, PPP)", 
        "GDP per Capita (PPP)", 
        "Unemployment (%)", 
        "FDI Inflow (Millions)"
        ]
    ]
teste.head()
```

```python
# Printing Information after transformation
print(teste.info())
teste.head()
```

```python
# Grouping and printing data
mean_unemployment_region\
    = t.group_data(
        economic_freedom_transformed, 
        "Region", 
        "Unemployment (%)",
        "count")

mean_unemployment_region.head(20)
```

```python
# Filtering and printing df
data_filtered = t.filter_data(
    economic_freedom_transformed, 
    'Region', 
    'Europe'
    )

data_filtered.head()
```

```python
# Total lines in a dataframe
total_rows_Europe = t.df_total_rows(data_filtered)
print(total_rows_Europe)
```

```python
# Rows having nan in a specific column
with_nan = t.show_where_is_null(economic_freedom_transformed, "FDI Inflow (Millions)")
with_nan
```

```python
# Deleting rows where the column have nan values
without_nan = t.delete_where_is_null(economic_freedom_transformed, "FDI Inflow (Millions)")
```

```python
# Comparing the number of rows before/after delete nan 
print(f"Row number with nan: {economic_freedom_transformed.shape[0]}")
print(f"Row number without nan: {without_nan.shape[0]}")
```
